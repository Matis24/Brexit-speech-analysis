```python
from bs4 import BeautifulSoup
import pandas as pd

# Charger le contenu du fichier
with open('/content/pressebrexit.xml.txt', 'r', encoding='utf-8') as file:
    content = file.read()

# Utiliser BeautifulSoup pour parser le contenu HTML/XML
soup = BeautifulSoup(content, 'html.parser')

# Trouver tous les éléments 'text' (ou tout autre balise selon votre structure)
texts = soup.find_all('text')

# Créer un DataFrame pour stocker les données extraites
df = pd.DataFrame(columns=['id', 'title', 'author', 'year', 'date', 'section', 'sourcename', 'text'])

# Extraire les informations nécessaires et les stocker dans le DataFrame
for text in texts:
    text_data = {
        'id': text.get('id'),
        'title': text.get('title'),
        'author': text.get('author'),
        'year': text.get('year'),
        'date': text.get('date'),
        'section': text.get('section'),
        'sourcename': text.get('sourcename'),
        # Extraire le texte proprement dit, en supposant qu'il est directement dans <text> ou vous pouvez adapter selon votre structure
        'text': text.get_text(separator=" ", strip=True)
    }
    df = df.append(text_data, ignore_index=True)
```
